```python
import random
```

```python
def fitness(x):
    return x**2

def generate_population(size):
    return [random.randint(0,10) for _ in range(size)]

def slection (population):
    population =sorted(population, key =fitness, reverse=True)
    return population[:2]

def crossover(parent1,parent2):
    point=random.randint(1,4)
    mask=(1<<point)-1
    child=(parent1 & mask) | (parent2 & ~mask)
    return child 

def mutation (child):
    bit=1<<random.randint(0,4)
    return child ^bit


def genetic_algorithm():
    population=generate_population(100)

    for generation in range (10):
        print(f"Generation {generation}: {population}")

        parent1, parent2 = slection(population)

        child = crossover(parent1, parent2)
        child = mutation(child)


        population[random.randint(0,99)] = child

    best=max(population, key=fitness)
    print(f"Best solution: {best} with fitness {fitness(best)}")
```

```python
genetic_algorithm()
```
