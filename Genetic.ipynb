```python
import random

def fitness(x):
    return -(x-5)**2 + 25   # peak at x=5

def genetic():
    population = [random.randint(0,10) for _ in range(6)]

    for _ in range(10):
        population = sorted(population, key=fitness, reverse=True)
        new = population[:2]

        while len(new) < 6:
            parent = random.choice(population[:3])
            child = parent + random.randint(-1,1)
            new.append(child)

        population = new

    return max(population, key=fitness)

print("\nGenetic Best:", genetic())
```
