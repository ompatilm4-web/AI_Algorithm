##Forward Chaining 


```python
#Forward Chaining
def forward_chaining(rules, facts, goal):
    inferred = set(facts)

    while True:
        new_inferred = set()

        for condition, result in rules:
            if all(c in inferred for c in condition) and result not in inferred:
                new_inferred.add(result)

        if not new_inferred:
            break

        inferred.update(new_inferred)

        if goal in inferred:
            return True, inferred

    return False, inferred


```

```python
# Example
rules = [
    (['A', 'B'], 'C'),
    (['C'], 'D'),
    (['D'], 'E')
]

facts = ['A', 'B']
goal = 'E'

result, inferred_facts = forward_chaining(rules, facts, goal)

print("Goal reached:", result)
print("Inferred facts:", inferred_facts)
```

Backward Chaining 

```python
def backward_chaining(rules, facts, goal, visited=None):
    if visited is None:
        visited = set()

    # If goal already known
    if goal in facts:
        return True

    # Avoid loops
    if goal in visited:
        return False

    visited.add(goal)

    # Find rules that conclude the goal
    for condition, result in rules:
        if result == goal:
            if all(backward_chaining(rules, facts, c, visited) for c in condition):
                return True

    return False


```

```python
# Example
rules = [
    (['A', 'B'], 'C'),
    (['C'], 'D'),
    (['D'], 'E')
]

facts = ['A', 'B']
goal = 'E'

result = backward_chaining(rules, facts, goal)

print("Goal reached:", result)
```
