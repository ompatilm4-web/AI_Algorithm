```python
# Forward Chaining

def forward_chaining(rules, facts):
    inferred = set(facts)

    while True:
        added = False

        for premise, conclusion in rules:
            if set(premise).issubset(inferred) and conclusion not in inferred:
                inferred.add(conclusion)
                added = True

        if not added:
            break

    return inferred


# Example
rules = [
    (['A', 'B'], 'C'),
    (['C'], 'D')
]

facts = ['A', 'B']

print("Inferred Facts:", forward_chaining(rules, facts))
```
