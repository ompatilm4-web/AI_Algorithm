```python
#Alpha Beta pruning implementation in python
import math
```

```python
def alpha_beta_pruning(Depth, nodeIndex, maximizingPlayer, values, alpha, beta,maxDepth):
    if Depth == maxDepth:
        return values[nodeIndex]
    

    if maximizingPlayer:
        best = -math.inf

        for i in range (2):
            val=alpha_beta_pruning(Depth+1, nodeIndex*2 + i, False, values,alpha,beta,maxDepth)
            best =max(best,val)
            alpha=max(alpha,best)
            if beta <=alpha :
                break
        return best
    else:
        best = math.inf

        for i in range (2):
            val=alpha_beta_pruning(Depth+1, nodeIndex*2 + i, True, values,alpha,beta,maxDepth)
            best =min(best,val)
            beta=min(beta,best)
            if beta <=alpha :
                break
        return best
    
```

```python
values = [3,5,2,9,12,1]

maxDepth = int(math.log2(len(values)))
print("Optimal Value:", alpha_beta_pruning(0, 0, True, values, -math.inf, math.inf, maxDepth))
```
