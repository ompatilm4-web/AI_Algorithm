```python
# Backward Chaining

def backward_chaining(goal, rules, facts):
    if goal in facts:
        return True

    for premise, conclusion in rules:
        if conclusion == goal:
            if all(backward_chaining(p, rules, facts) for p in premise):
                return True

    return False


# Example
rules = [
    (['A', 'B'], 'C'),
    (['C'], 'D')
]

facts = ['A', 'B']

print("Can we prove D?", backward_chaining('D', rules, facts))
```
