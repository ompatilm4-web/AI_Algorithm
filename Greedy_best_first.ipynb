```python
# Greedy first search algorithm implementation
from queue import PriorityQueue

graph ={
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 3)],
    'D': [('G', 1)],
    'E': [('H', 2)],
    'F': [('I', 1)],
    'G': [('J', 1)],
    'H': [('K', 1)],
    'I': [('L', 1)],
    'J': [('M', 1)],
    'K': [('N', 1)],
    'L': [('O', 1)],
    'M': [('P', 1)],
    'N': [('Q', 1)],
    'O': [('R', 1)],
    'P': [('S', 1)],
    'Q': [('T', 1)],
    'R': [('U', 1)],
    'S': [('V', 1)],
    'T': [('W', 1)],
    'U': [('X', 1)],
    'V': [('Y', 1)],
    'W': [('Z', 1)]
}
```

```python
def greedy_best_first_search(graph, start, goal):
    visited = set()
    queue = PriorityQueue()
    queue.put((0, start))  # (priority, node)

    while not queue.empty():
        _, current_node = queue.get()

        if current_node == goal:
            return True  # Goal found

        if current_node not in visited:
            visited.add(current_node)

            for neighbor, cost in graph.get(current_node, []):
                if neighbor not in visited:
                    queue.put((cost, neighbor))  # Add neighbor to the priority queue

    return False  # Goal not found
```

```python
# Example usage
start_node = 'A'
goal_node = 'Z'
if greedy_best_first_search(graph, start_node, goal_node):
    print(f"Path from {start_node} to {goal_node} found.")
else:
    print(f"Path from {start_node} to {goal_node} not found.")

    
```
