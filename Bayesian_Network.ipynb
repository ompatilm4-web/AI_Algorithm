```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
```

```python
#Define the Structure 
model = DiscreteBayesianNetwork([
    ('Rain', 'Sprinkler'),
    ('Rain', 'WetGrass'),
    ('Sprinkler', 'WetGrass')

])
```

```python
#define the CPDs

#Rain 
cpd_rain = TabularCPD(
    variable='Rain',
    variable_card=2,
    values=[[0.2], [0.8]]
)

#Sprinkler
cpd_sprinkler = TabularCPD(
    variable='Sprinkler',
    variable_card=2,
    values=[[0.01, 0.4], [0.99, 0.6]],
    evidence=['Rain'],
    evidence_card=[2]
)

#WetGrass
cpd_WetGrass =TabularCPD(
    variable='WetGrass',
    variable_card=2,
    values=[[0.8, 0.9, 0.1, 0.2], [0.2, 0.1, 0.9, 0.8]],
    evidence=['Rain', 'Sprinkler'],
    evidence_card=[2, 2]
)

#Add CPDs to the model
model.add_cpds(cpd_rain, cpd_sprinkler, cpd_WetGrass)
```

```python
#Check Model Validity
print(model.check_model())
#Inference
inference = VariableElimination(model)
#P(Rain | WetGrass = True)
query_result = inference.query(variables=['Rain'], evidence={'WetGrass': 1})
print(query_result)
```
