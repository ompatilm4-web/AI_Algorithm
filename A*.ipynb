```python
import heapq

def a_star(graph, start, goal, heuristic):
    pq = [(0, start)]
    g_cost = {start: 0}
    parent = {start: None}

    while pq:
        _, node = heapq.heappop(pq)

        if node == goal:
            path = []
            while node:
                path.append(node)
                node = parent[node]
            return path[::-1]

        for neighbor, cost in graph[node]:
            new_cost = g_cost[node] + cost

            if neighbor not in g_cost or new_cost < g_cost[neighbor]:
                g_cost[neighbor] = new_cost
                f = new_cost + heuristic[neighbor]
                heapq.heappush(pq, (f, neighbor))
                parent[neighbor] = node

    return []

graph_w = {
    'A':[('B',1),('C',4)],
    'B':[('D',2),('E',5)],
    'C':[('F',1)],
    'D':[], 'E':[], 'F':[]
}

heuristic = {'A':7,'B':6,'C':2,'D':1,'E':0,'F':0}

print("\nA* Path:", a_star(graph_w, 'A', 'F', heuristic))
```
